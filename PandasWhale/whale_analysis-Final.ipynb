# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp500_history.csv`: Contains historical closing prices of the S&P 500 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data

```python
# Reading whale returns
#nflx=pd.read_csv(filePathNFLX,index_col="date", infer_datetime_format=True, parse_dates=True).dropna()#
whale_returns = pd.read_csv(filepathWhale,index_col='Date',infer_datetime_format=True,parse_dates=True)
```

```python
# Count nulls
print (whale_returns.isna().count())
```

```python
# Drop nulls
whale_returns.dropna()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data

```python
# Reading algorithmic returns
algo_returns = pd.read_csv(filepathAlgo,index_col='Date',infer_datetime_format=True,parse_dates=True)
```

```python
# Count nulls
print (algo_returns.isna().count())
```

```python
# Drop nulls
algo_returns.dropna()
```

## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
SP_history = pd.read_csv(filepathSP,index_col='Date',infer_datetime_format=True,parse_dates=True)
```

```python
# Check Data Types
SP_history.dtypes
```

```python
# Fix Data Types
SP_history['Close']=SP_history['Close'].str.replace('$','',regex=False)
SP_history['Close']=SP_history['Close'].astype(float)
SP_history.dtypes
```

```python
# Calculate Daily Returns
SP_returns=SP_history.pct_change()
```

```python
# Drop nulls
SP_returns.dropna()
```

```python
# Rename `Close` Column to be specific to this portfolio.
SP_returns=SP_returns.rename(columns={'Close':'SP500'}).dropna()
```

## Combine Whale, Algorithmic, and S&P 500 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
combined_returns=pd.concat([whale_returns,algo_returns,SP_returns], axis="columns", join="inner").dropna()
combined_returns.head()
```

```python
# Calculate the correlation
# Display de correlation matrix
correllation=combined_returns.corr()
sns.heatmap(correllation,vmin=-.5,vmax=.5)

```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
#rolling_covariance = daily_returns['AMZN'].rolling(window=30).cov(daily_returns['S&P 500'])
rolling_covariance=combined_returns['Algo 1'].rolling(window=30).cov(combined_returns['SP500'])
rolling_covariance
# Calculate variance of S&P 500
rolling_variance=combined_returns['SP500'].rolling(window=30).var()
rolling_variance
# Computing beta
a_beta=rolling_covariance/rolling_variance
a_beta
# Plot beta trend
a_beta.plot(figsize=(20,10))
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratio= (combined_returns.mean() -0) / (combined_returns.std() * np.sqrt(252))
sharpe_ratio.head()
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio.plot(kind='bar', figsize=(20,10))
```

## Choose 3-5 custom stocks with at last 1 year's worth of historic prices and create a DataFrame of the closing prices and dates for each stock.

```python
# Reading data from 1st stock
df_TQQQ = pd.read_csv(filepathTQQQ,index_col='Date',infer_datetime_format=True,parse_dates=True)
df_TQQQ.columns = ['TQQQ']
df_TQQQ.head()
```

```python
# Reading data from 2nd stock
df_TMF = pd.read_csv(filepathTMF,index_col='Date',infer_datetime_format=True,parse_dates=True)
df_TMF.columns = ['TMF']
df_TMF.head()
```

```python
# Reading data from 3rd stock
df_SCHD = pd.read_csv(filepathSCHD,index_col='Date',infer_datetime_format=True,parse_dates=True)
df_SCHD.columns = ['SCHD']
df_SCHD.head()
```

```python
# Combine all stocks in a single DataFrame
df_custom_portfolio=pd.concat([df_TQQQ,df_TMF,df_SCHD], axis="columns", ignore_index=True).dropna()
df_custom_portfolio.columns = ["TQQQ","TMF","SCHD"]
df_custom_portfolio.head()
```

```python
# Calculate daily returns & # Drop NAs
custom_portfolio_returns = df_custom_portfolio.pct_change().dropna()

# Display sample data
custom_portfolio_returns.head()
```

## Calculate the weighted returns for the portfolio assuming an equal number of shares for each stock

```python
# Set weights
weights = [1/3, 1/3, 1/3]

# Calculate portfolio return
returns = custom_portfolio_returns.dot(weights)
returns.columns = ["My_Portfolio"]
# Display sample data
returns.head()
```

## Join your portfolio returns to the DataFrame that contains all of the portfolio returns

```python
# Join your returns DataFrame to the original returns DataFrame
all_returns=pd.concat([combined_returns,returns], axis="columns", ignore_index=True).dropna()
all_returns.columns = ["SOROS FUND MANAGEMENT LLC","PAULSON & CO.INC.","TIGER GLOBAL MANAGEMENT LLC","BERKSHIRE HATHAWAY INC","Algo 1","Algo 2","SP500","Custom"]
all_returns.head()
```

## Re-run the risk analysis with your portfolio to see how it compares to the others

### Calculate and Plot Rolling 60-day Beta for Your Portfolio compared to the S&P 500
